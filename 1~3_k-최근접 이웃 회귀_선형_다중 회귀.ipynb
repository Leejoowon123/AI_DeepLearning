# 정리

***Summary***
  - 특성 : 데이터를 표현하는 하나의 성질
  - 훈련 : 머신러닝 알고리즘이 데이터에서 규칙을 찾는 과정
  - 모델 : 알고리즘이 구현된 객체
  - 정확도 : 사이킷런에서는 0 ~ 1사이 (클수록 정확함)
  - 지도학습 : 입력과 타깃을 전달해 모델을 훈련한 다음 새로운 데이터를 예측하는데 활용
  - 비지도 학습: 타깃 데이터가 없이, 입력 데이터에서 어떤 특성을 찾는 데 주로 이용
  - 훈련 세트 : 모델 훈련 시 사용 데이터
  - 테스트 세트 : 전체 데이터의 20 ~ 30% 정도 사용. 데이터가 클 경우 1%로도 충분
  - 데이터 전처리 : 머신 러닝 모델에 훈련 데이터를 주입 전 데이터를 가공하는 단계
  - 표준 점수: 훈련 세트의 스케일을 바꾸는 방법(특성의 평균을 빼고 표준편차로 나눔)  **반드시 훈련세트의 평균과 표준편차로 테스트 세트를 바꿔야함**
  - 브로드캐스팅 : 크기가 다른 넘파이 배열에서 자동으로 사칙 연산을 모든 행이나 열로 확장하여 수행하는 기능
  - 회귀: 임의의 수치를 예측하는 문제.따라서 타깃값도 임의의 수치가 됨
  - **k-최근접 이웃 회귀**: k-최근접 이웃 알고리즘을 사용해 회귀문제를 품. (가장 가까운 이웃 샘플을 찾고, 샘플들의 타깃값을 평균해 예측으로 삼음)
  - 결정계수(R^2): 대표적인 회귀 문제의 성능 측정 도구. 0 ~ 1 (1에 가까울 수록 좋음)
  - 과대적합: 훈련 세트 성능 > 테스트 세트 성능.(내재된 거시적인 패턴 감지 불가)
  - 과소 적합: 과대 적합과 반대이거나 훈련세트&&테스트세트 모두 동일하게 낮은 경우 ==> 복잡한 모델을 사용해 훈련 세트에 맞는 모델을 생성해야함.
  - **선형 회귀** : 특성과 타깃 사이의 관계를 가장 잘 나타내는 선형 방정식을 찾는 방법
  - 선형회귀가 찾은 특성과 타깃 사이의 관계 = 선형방정식의 계수|가중치에 저장  
          *참고:머신러닝에서 가중치는 종종 방정식의 기울기와 절편을 모두 의미하는 경우가 多
  - **모델 파라미터**: 선형회귀가 찾은 가중치를 머신러닝 모델이 특성에서 학습한 파라미터
  - **다항 회귀**: 다항식을 사용해 특성과 타깃 사이의 관계를 나타냄.
  - **다중 회귀**: 여러개의 특성을 사용하는 회귀 모델
  - 특성공학: 주어진 특성을 조합해 새로운 특성을 만드는 일련의 작업
  - **릿지** : 규제가 있는 선형회귀 모델. 선형모델의 계수를 작게해 과대적합을 완화
  - **라쏘** : 규제가 있는 선형회귀 모델. 릿지와 달리 계수 값을 아예 0으로 만들어버리는 경우가 있음
  - **하이퍼파라미터**: 머신러닝 알고리즘이 학습하지 않은 파라미터. (사람이 사전에 지정해줘야함)
                  대표적으로 릿지와 라쏘의 규제강도인 alpha

**matplotlib 패키지**
  - **scatter()** : 산점도를 그리는 함수


**scikit-learn**
  - **KNeighborsClassifier()** : **k-최근접 이웃 분류 모델**을 만드는 사이킷런 클래스
  - **fit()** : 사이킷런을 훈련할 때 사용하는 매서드, 처음 두 매개변수로 훈련에 사용할 특성과 정답 데이터 전달
  - **predict()** : 예측할 때 사용하는 메서드. 특성 데이터 하나만 매개변수로 받음
  - **score( )** : 모델의 성능 측정 메서드. predict()로 예측 수행 후 분류 모델일 경우 정답과 비교하여 올바르게 예측한 비율을 반환
  - **train_test_split()**: 훈련 데이터&테스트 데이터를 나누는 함수 비율 은 **test_size**에서 조절 가능. default=0.25(25%)
                              **stratify**매개변수에 클래스 레이블이 담긴 배열을 전달하면 클래스 비율에 맞게 나누어 줌
  - **kneighbors()**: k-최근접 이웃 객체의 메서드. 입력 데이터에 가장 가까운 이웃을 찾아 거리와 이웃 샘플의 인덱스를 반환
  - **KNeighborsRegressor()**: **k-최근접 이웃 회귀 모델 생성**하는 클래스. n_neighbors 매개변수로 이웃의 개수를 지정(default=5)
  - **mean_absolute_error()** : 회귀 모델의 평균 절댓값 오차를 계산. 매개변수(타깃, 예측값) 전달.
  - **mean_squared_error()** : 평균 제곱 오차 계산
  - **LinearRegression()**: **선형회귀** 클래스
  - **PolynomialFeatures**: 주어진 특성을 조합해 새로운 특성 생성. degree=최고 차수 지정(default: 2)
              interaction_only가 True 일경우 : 거듭제곱 항은 제외 & 특성 간의 곱셈항만 추가 (default: False)
              include_bias가 False 인 경우 : 절편을 위한 특성을 추가하지 않음 (default: True)
  - **Ridge** : 릿지 회귀모델 훈련. alpha가 커질수록 규제가 세짐(default: 1)
  - **Lasso** : 라쏘 회귀모델 훈련. alpha와 random_state 매개변수는 Ridge 클래스와 동일. max_iter: 알고리즘 수행 반복횟수 지정(default: 1000)


**numpy 패키지**
  - **seed()** : 난수를 생성하기 위한 정수 초깃값 지정
  - **arange()** : 일정 간격의 정수 또는 실수 배열 생성
  - **shuffle()** : 주어진 배열을 랜덤하게 섞음. 다차원 배열의 경우 첫 번째 축(행)에 대해서만 섞음.
  - **reshape()** : 배열의 크기를 바꾸는 메소드.

```python
import matplotlib.pyplot as plt
```

지도 학습은 입력(데이터)와 타깃(정답)으로 이뤄진 훈련 데이터가 필요하다.

**특성 : 도미와 빙어의 length, weight **

---



```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show
```

```python
#빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
#합친 내용 출력
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 첫 번째 머신러닝 프로그램
lenght = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l,w] for l, w in zip(lenght, weight)]
```

```python
print(fish_data)
```

```python
fish_target = [1] * 35 + [0] * 14 #도미 = 1, 방어 = 0
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNCF
#사이킷런 패키지에서 k-최근접 이웃 알고리즘 구현 클래스 : KNeighborClassifier
kn = KNCF()
```

```python
kn.fit(fish_data, fish_target) #알고리즘 훈련
kn.score(fish_data, fish_target) #score반환값 : 0 ~ 1 높을수록 정확도 높음
```

```python
kn.predict([[30,600]]) #predict(): 훈련하고 예측할 때 사용. 특성 데이터 하나만 매개변수로 받음
kn49 =KNCF(n_neighbors=49) #이렇게 하면 49개 중 도미(0)이 35개이므로 무엇을 넣어도 0으로 예측
kn49.fit(fish_data, fish_target) #fit(): 모델 훈련 처음 두 매개변수로 훈련에 사용할 특성 + 정답 데이터 전달
kn49.score(fish_data, fish_target) #score(): 모델 성능 측정
```

```python
# k-최근접 이웃 알고리즘
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
#정확도 높이기
for n in range(5, 50):
  #k-최근접 이웃 개수 설정
  kn.n_neighbors = n
  score = kn.score(fish_data, fish_target)
  if score < 1:
    print(n, score)
    break
```

# 샘플링편향 해결

현재 훈련세트(35개)와 테스트 세트(14개)로 나눔
훈련세트에는 도미가 다 들어있고
테스트 세트에는 빙어가 다 들어있어서 '0'이 출력함.

==> 이러한 현상: 샘플링 편향

```python
import numpy as np
```

생선데이터를 2차워 넘파이 배열로 변환

```python
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)
```

```python
np.random.seed(42) #책과 일정한 결과를 얻을 수 있도록 seed(정수 초깃값: 초깃값 같으면 동일 난수 출력 가능)를 지정
index = np.arange(49)
np.random.shuffle(index) #shuffle로 인덱스 섞기
print(index)
```

```python
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
plt.scatter(train_input[:,0], train_input[:,1]) #2차원 배열 So, 행과 열 index를 , 나누어 지정
plt.scatter(test_input[:,0], test_input[:,1]) # So, : , 0 이렇게 됨
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# fit()을 하면 이전에 했던 내용 사라짐 So, 유지하고 싶으면 새로운 객체를 생성해야함
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
test_target #코랩은 마지막 코드 결과 자동으로 출력 So, print 필요 x
```

**두 리스트를 연결(튜플로) 튜플은 수정 불가하기에 믿을 수 있어서 매개변수로 자주 사용**

```python
fish_data = np.column_stack((fish_length,fish_weight))
```

원하는 개수의 1과 0을 채운 배열을 만들어줌

**- np.ones()**

**- np.zeros()**

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
```

```python
from sklearn.model_selection import train_test_split
```

train_test_split() 함수: 25%를 테스트 세트로 떼어냄

```python
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42)
```

```python
print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
print(test_target)
```

test_target이 도미(10개), 빙어(3개)로 샘플링편향이 나타남

train_test_split()함수에서 stratify 매개변수에 타깃 데이터를 전달하면 해결 가능

```python
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state = 42)
```

```python
kn = KNCF()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
# 오류가 발생한 부분을 ^로 표시
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexs = kn.kneighbors([[25,150]])
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(train_input[indexs,0], train_input[indexs, 1], marker='D')
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

눈으로 보면 가까운 도미(위에 있는 점들)가 근접한대 왜 이웃점은 빙어로 예측을 할까?

x, y축의 기준이 다르기 때문. So, 기준을 맞춰줘야함
:(이 과정을 데이터 전처리 과정이라고 함)

데이터 전처리 방법:
1. 표준점수(각 특성값이 평균에서 표준편차의 몇 배만큼 떨어져있는 지 나타냄)
  
  ==> 실제 특성값의 크기와 상관없이 동일한 조건으로 비교가 가능

표준점수 변환 = (원본데이터-평균)/표준편차

```python
mean = np.mean(train_input, axis=0) #평균계산
std = np.std(train_input, axis=0) #표준편차 계산
```

```python
train_scaled = (train_input - mean)/std
```

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

샘플도 동일한 비율로 변환해줘야함

**<중요> 훈련세트의 mean, std를 이용해서 변환해야 함**

**2차 방정식의 그래프가 필요(다항회귀)**

```python
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input**2, test_input))
```

```python
lr = LR()
lr.fit(train_poly, train_target)
print("예측값: ",lr.predict([[50**2, 50]]))
print("훈련계수&절편: ", lr.coef_, lr.intercept_)
```

# 규제를 한 릿지&라쏘 분석

특성을 늘려서 훈련 세트에 점수를 높일 수 있다

하지만 너무 높아지면 과대 적합되어 테스트 세트에서 점수가 낮게 나오는 경우가 발생한다!!!

So **규제**(과대한 학습을 막기위해 훼방을 하는것 = 과대적합을 막는것)가 필요!

**릿지(ridge)&라쏘(lasso)** : 선형회귀 모델에 규제를 추가한 모델
  - 릿지: 계수를 제곱한 값을 기준으로 규제 적용
  - 라쏘 :계수의 절댓값을 기준으로 규제 적용

선형회귀 모델의 경우 : 특성에 곱해지는 ** 계수(|기울기)의 크기를 줄여주는 것**

```python
from sklearn.preprocessing import StandardScaler as SS
ss = SS()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_scaled, train_target)
print('릿지 훈련: ', ridge.score(train_scaled, train_target))
print('릿지 테스트: ', ridge.score(test_scaled, test_target))
```

릿지, 라쏘 모델 사용 시 규제의 양을 임의로 조절이 가능하다.
  - 모델 객체를 만들 때 alpha 매개변수로 규제의 강도를 조절
  - alpha 값이 큼 --> 규제 강도 세짐 --> 계수 값을 줄임 --> 조금 더 과소적합되도록 유도
  - alpha 값이 작음 --> 계수를 줄이는 역할 감소 --> 선형 회귀모델과 유사 --> 과대적합될 가능성이 높음

**하이퍼파라미터**: 릿지모델이 학습하는 값이 아니라 사전에 우리가 지정해야 하는 값
  - 적절한 alpha값 찾기 : **alpha 값에 대한 R^2값의 그래프를 그려보는 것**

```python
train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list:
  ridge = Ridge(alpha = alpha)
  ridge.fit(train_scaled, train_target)
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score) #위에 파란색 그래프
plt.plot(np.log10(alpha_list), test_score) #아래 주황색 그래프
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

**alpha 값을 0.001 부터 10배씩 늘렸기 때문에 이대로 그리면 그래프 왼쪽이 너무 촘촘해짐**
  - So, alpha_list에 있는 6개 값을 동일 한 간격으로 나타내기 위해 **로그 함수**로 바꾸어 지수로 표현
  - ex) 0.001 = -3, 0.01 = -2

넘파이의 로그함수 = **np.log()**: 자연상수 e를 밑으로 하는 자연로그 & **np.log10()** : 상용로그

**결론:** 적절한 alpha 값 = 두 그래프가 가장 가깝고, 테스트 세트의 점수가 가장 높은 -1(0.1) ==> **alpha 값 = 0.1**로 최종 모델 훈련

```python
ridge = Ridge(alpha = 0.1)
ridge.fit(train_scaled, train_target)
print('릿지 훈련 점수: ',ridge.score(train_scaled, train_target))
print('릿지 테스트 점수: ', ridge.score(test_scaled, test_target))
```

**` 라쏘 회귀 `**



```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print('라쏘 훈련 점수: ', lasso.score(train_scaled, train_target))
print('라쏘 테스트 점수 ', lasso.score(test_scaled, test_target))
```

```python
train_score, test_score = [], []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  lasso = Lasso(alpha = alpha, max_iter = 10000) #반복횟수 값 : max_iter
  lasso.fit(train_scaled, train_target)
  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()

lasso = Lasso(alpha = 10)
lasso.fit(train_scaled, train_target)
print('라쏘 훈련 점수: ', lasso.score(train_scaled, train_target))
print('라쏘 테스트 점수: ', lasso.score(test_scaled, test_target))
```

특성을 많이 사용했지만, 릿지와 마찬가지로 라쏘 모델이 과대적합을 잘 억제하고 테스트의 성능을 크게 높여줌

라쏘 모델의 계수는 coef_ 속성에 저장. 그리고 **라쏘 모델은 계수 값이 아예 0으로 만들어 질수도 있음**
